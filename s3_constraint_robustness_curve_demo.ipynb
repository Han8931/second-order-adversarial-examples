```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import sklearn.metrics

# import matplotlib
# matplotlib.rc('text', usetex=True)

sns.set()
sns.set_style('whitegrid')

fig, ax = plt.subplots(figsize=(10,8))

lw = 4 # line width


x_factor = 3 + 1/3
y_factor = 1 / .7
x = np.arange(0, 1, 0.001) / x_factor

y1 = ((x * x_factor) ** (10)) / y_factor # y = x
y2 = ((x * x_factor) ** .1) / y_factor
y3 = ((x * x_factor)) / y_factor

auc1 = sklearn.metrics.auc(x, y1)
accs1 = auc1 * (x_factor * y_factor)
sns.lineplot(x, y1, ax=ax, linewidth=lw, label=f'Worthless constraint (ACCS = {accs1:.3f})') # y = x


auc2 = sklearn.metrics.auc(x, y2)
accs2 = auc2 * (x_factor * y_factor)
sns.lineplot(x, y2, ax=ax, linewidth=lw, label=f'Excellent constraint (ACCS = {accs2:.3f})')


auc3 = sklearn.metrics.auc(x, y3)
accs3 = auc3 * (x_factor * y_factor)
sns.lineplot(x, y3, ax=ax, linewidth=lw, label=f'Good constraint (ACCS = .500)')


ax.get_lines()[-3].set_linestyle('--')
ax.get_lines()[-1].set_linestyle(':')

plt.xlabel('Adversarial Examples in Constraint', fontsize=18)
plt.ylabel('Adversarial Examples in Attacked Model', fontsize=18)
plt.title('Constraint Robustness Curve (Example)', fontsize=22)
plt.legend(loc='lower right', prop={'size': 18}) # Adjust legend font size

plt.tight_layout()
plt.savefig('accs_example.pdf')
```
